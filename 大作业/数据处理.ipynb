```python
# -*- coding: utf-8 -*-
"""
UNESCO项目数据分析
数据挖掘大作业
"""

# 导入必要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
#基本数据清洗
# -*- coding: utf-8 -*-
"""
读取UNESCO项目数据并过滤掉支出金额为0的记录
"""

import pandas as pd
import os

# 文件路径
file_path = 'UNESCO-projects-prj001.csv'

print(f"正在读取文件: {file_path}")

# 检查文件是否存在
if not os.path.exists(file_path):
    print(f"错误: 找不到文件 {file_path}")
    exit(1)

# 读取CSV文件
df = pd.read_csv(file_path)
print(f"原始数据形状: {df.shape}")

# 检查是否有Project Expenditure Amount列
if 'Project Expenditure Amount' not in df.columns:
    print("错误: 文件中没有 'Project Expenditure Amount' 列")
    print(f"可用的列: {list(df.columns)}")
    exit(1)

# 转换为数值类型（如果不是）
df['Project Expenditure Amount'] = pd.to_numeric(df['Project Expenditure Amount'], errors='coerce')

# 过滤掉支出金额为0或NaN的记录
filtered_df = df[df['Project Expenditure Amount'] > 0].copy()
print(f"过滤后数据形状: {filtered_df.shape}")
print(f"过滤掉的记录数: {len(df) - len(filtered_df)}")

# 保存过滤后的数据
output_file = 'UNESCO_projects_filtered.csv'
filtered_df.to_csv(output_file, index=False)
print(f"过滤后的数据已保存到: {output_file}")

# 显示一些基本统计信息
print("\n过滤后数据的支出统计:")
print(f"最小支出: {filtered_df['Project Expenditure Amount'].min()}")
print(f"最大支出: {filtered_df['Project Expenditure Amount'].max()}")
print(f"平均支出: {filtered_df['Project Expenditure Amount'].mean():.2f}")
print(f"中位数支出: {filtered_df['Project Expenditure Amount'].median():.2f}")
```
